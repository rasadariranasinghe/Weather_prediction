## Creating the model

```python
# initilaizing the model
from sklearn.linear_model import Ridge
model = Ridge(alpha=0.1)
```

```python
train_data = weather_core_df.loc[:'1985-01-01']
test_data = weather_core_df.loc['1985-01-02':]
```

```python
train_data.tail()
```

```python
test_data.head()
```

```python
# Select features and target for training data
X_train = train_data[['Latitude', 'Longitude', 'Elevation', 'Precipitation', 'Snow' ,'Snow_Depth','Temp_Min']]
y_train = train_data['Temp_Max']

# Select features and target for testing data
X_test = test_data[['Latitude', 'Longitude', 'Elevation', 'Precipitation', 'Snow','Snow_Depth', 'Temp_Min']]
y_test = test_data['Temp_Max']
```

```python
# Train the Ridge regression model
model.fit(X_train, y_train)
```

```python
# Predict on the testing data
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Evaluate the model using Mean Squared Error (MSE) and R-squared (R²) metrics
# These metrics indicate the accuracy of the predictions
mse_temp_max = mean_squared_error(y_test, y_pred)
r2_temp_max = r2_score(y_test, y_pred)

# Print the evaluation metrics
print(f"Temp_Max - MSE: {mse_temp_max}, R²: {r2_temp_max}")
```

```python
# Visualize residuals to analyze prediction errors
residuals = y_test - y_pred
residuals.plot(title="Residuals for Temp_Max Prediction")
plt.ylabel('Residual Value')
plt.show()
```

```python
# Convert predictions to a DataFrame
y_pred_df = pd.DataFrame(y_pred, columns=['Predicted_Temp_Max'], index=y_test.index)

# Combine actual and predicted values into a single DataFrame
comparison_df = pd.concat([y_test, y_pred_df], axis=1)

# Display the comparison DataFrame
comparison_df.head()
```

```python
model.coef_
```

```python
# Compare actual vs predicted values using a scatter plot
# This helps to visually assess how well the model is performing
comparison_df.plot(kind='scatter', x='Temp_Max', y='Predicted_Temp_Max', title='Actual vs Predicted Temp_Max')
plt.xlabel('Actual Temp_Max')
plt.ylabel('Predicted Temp_Max')
plt.show()

# Plot time-series data for actual and predicted Temp_Max values
comparison_df.plot(title='Temp_Max: Actual vs Predicted (Time Series)')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
# Analyze feature importance by extracting coefficients from the trained Ridge model
# This shows which features contribute most to predicting Temp_Max
feature_importance = pd.Series(model.coef_, index=X_train.columns)

# Plot feature importance as a bar chart for better visualization
feature_importance.plot(kind='bar', title='Feature Importance for Temp_Max Prediction')
plt.ylabel('Coefficient Value')
plt.show()
```

```python
# Train a Linear Regression model for comparison
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Predict Temp_Max using the Linear Regression model
y_pred_lr = lin_reg.predict(X_test)

# Evaluate and compare the R² scores of both models
print("Linear Regression R²:", r2_score(y_test, y_pred_lr))
print("Ridge Regression R²:", r2_temp_max)

# Plot comparison between Ridge and Linear Regression predictions
plt.plot(y_test.index, y_test, label='Actual', color='black')
plt.plot(y_test.index, y_pred, label='Ridge Predictions', color='blue')
plt.plot(y_test.index, y_pred_lr, label='Linear Regression Predictions', color='red')
plt.title('Model Comparison: Ridge vs Linear Regression')
plt.legend()
plt.show()
```

```python
# Summarize key findings from the model
print("Key Findings:")
print("- Ridge Regression outperformed Linear Regression in predicting Temp_Max.")
print(f"- Important features influencing Temp_Max: {feature_importance.sort_values(ascending=False).index[:3].tolist()}")
print(f"- R² for Ridge Regression: {r2_temp_max:.2f}, indicating a strong model fit.")
print("- Visualizations confirm that predictions closely align with actual values over time.")
```
